SPY 計算ACF , PACF

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# 准备数据
data = pd.read_csv('ETHUSD.csv')  # 从CSV文件或其他数据源加载时间序列数据
series = data['Adj Close']  # 获取时间序列数据

# 绘制ACF图
sm.graphics.tsa.plot_acf(series, lags=20)  # 设置lags参数来指定要绘制的滞后阶数
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.title('Autocorrelation Function (ACF)')
plt.show()

# 绘制PACF图
sm.graphics.tsa.plot_pacf(series, lags=14)  # 设置lags参数来指定要绘制的滞后阶数
plt.xlabel('Lag')
plt.ylabel('PACF')
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA

# 讀取人口數據集
data = pd.read_csv('ETHUSD.csv')

# 將時間列設置為索引並轉換為時間序列
data['Date'] = pd.to_datetime(data['Date'])
#data.set_index('Year', inplace=True)

# 繪製人口數時間序列
data['Close'].plot()
plt.xlabel('Date')
plt.ylabel('Close')
plt.show()

# 擬合ARMA模型
#model = sm.tsa.ARMA(data['Born'], order=(4, 3)).fit()
model = ARIMA(data['Close'], order=(2, 1, 2))
model_fit = model.fit()
# 預測未來人口數
forecast = model_fit.forecast(steps=50)

# 繪製原始數據和預測結果
plt.plot(data['Close'], label='Actual')
plt.plot(forecast, label='Forecast')
#plt.xticks(years[::5])  # 设置刻度间隔为5年
#plt.xticklabels(years[::5])  # 设置刻度标签为每5年的年份值


plt.xlabel('Year')
plt.ylabel('Born')
plt.legend()
plt.show()
#print(model_fit.summary())

coefficient = model_fit.params
#print(coefficient)

print(model_fit.aic,model_fit.bic,model_fit.hqic)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg

data = pd.read_csv('DATA.csv')
data['Year'] = pd.to_datetime(data['Year'], format='%Y')
data.set_index('Year', inplace=True)

train_data = data.iloc[:-3]  # 使用除了最後三年的資料作為訓練集
train_data = train_data['Born']
test_data = data.iloc[-3:]  # 使用最後三年的資料作為測試集
model = AutoReg(train_data, lags=4)  # 設定lags參數来指定滯後階數
model_fit = model.fit()
predictions = model_fit.predict(start=len(train_data), end=len(train_data)+len(test_data)-1)
plt.plot(data.index, data['Born'], label='Actual')
plt.plot(test_data.index, predictions, label='Predicted')
plt.xlabel('Year')
plt.ylabel('Born')
plt.title('AR Model - Born Prediction')
plt.legend()
plt.show()
print(model_fit.aic,model_fit.bic,model_fit.hqic)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.vector_ar.var_model import VAR

data = pd.read_csv('DATA.csv')
data['Year'] = pd.to_datetime(data['Year'])
data.set_index('Year', inplace=True)
train_data = data.iloc[:-3]  # 使用除了最后三个时间点的数据作为训练集
test_data = data.iloc[-3:]  # 使用最后三个时间点的数据作为测试集
model = VAR(train_data)
model_fit = model.fit()
lag_order = model_fit.k_ar
input_data = train_data.values[-lag_order:]  # 最后lag_order个观测值作为输入数据
predictions = model_fit.forecast(y=input_data, steps=len(test_data))
predictions_df = pd.DataFrame(predictions, index=test_data.index, columns=test_data.columns)
plt.plot(data.index, data['Born'], label='Actual')
plt.plot(predictions_df.index, predictions_df['Born'], label='Predicted')
plt.xlabel('Year')
plt.ylabel('Born')
plt.title('VAR Model - Population Prediction')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sktime.forecasting.arima import ARIMA
from sktime.forecasting.base import ForecastingHorizon
import matplotlib.pyplot as plt

# 读取数据文件
data = pd.read_csv('DATA.csv')

# 转换为日期索引
data['Year'] = pd.to_datetime(data['Year'])
data = data.set_index('Year')

# 准备时间序列数据
y_train = data['Born']
exog_train = data[['GDP', 'Growth', 'Spend', 'NTDtoUSD']]

# 创建exog-ARMA模型对象
model = ARIMA(order=(3, 0, 2), seasonal_order=(0, 0, 0), suppress_warnings=True)

# 拟合exog-ARMA模型
model.fit(y_train, exog=exog_train)

# 定义预测时间范围
forecast_horizon = ForecastingHorizon(pd.date_range(start='2023-01-01', periods=5, freq='A'))

# 进行预测
y_pred = model.predict(forecast_horizon, exog=exog_train[-5:])

# 打印预测结果
print("预测结果:")
print(y_pred)

# 可视化预测结果
plt.figure(figsize=(10, 6))
plt.plot(data.index, y_train, label='真实值')
plt.plot(y_pred.index, y_pred, label='预测值')
plt.xlabel('年份')
plt.ylabel('出生人数')
plt.title('VARMAX模型预测出生人数')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
# 讀取人口數據集
data = pd.read_csv('DATA.csv')

# 將時間列設置為索引並轉換為時間序列
data['Year'] = pd.to_datetime(data['Year'])
#data.set_index('Year', inplace=True)

# 繪製人口數時間序列
data['Born'].plot()
plt.xlabel('Year')
plt.ylabel('Born')
plt.show()

p = False
n=0
while p == False:
    diff1 = data.diff(n)
    #diff1['Born'].plot()
    data_new = diff1['Born'].iloc[n:]
    result = adfuller(data_new)

    # 提取ADF檢定結果中的關鍵信息
    adf_statistic = result[0]
    p_value = result[1]
    critical_values = result[4]

    # 進行恆定性判斷
    if p_value < 0.05 and adf_statistic < critical_values['5%']:
        print("序列是恆定序列")
        p = True
        print(n)
    else:
        print("序列不是恆定序列")
        n = n+1
```
